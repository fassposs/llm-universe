测试 RunnableBranch的分支运行逻辑

```python
from langchain_core.runnables import RunnableBranch, RunnableLambda

# 创建分支路由器
branch = RunnableBranch(
    # 分支1: 如果输入是数字
    (lambda x: isinstance(x, (int, float)), 
     RunnableLambda(lambda x: f"数字: {x}")),
    
    # 分支2: 如果输入是字符串
    (lambda x: isinstance(x, str), 
     RunnableLambda(lambda x: f"字符串: {x}")),
    
    # 默认分支（最后一个参数）
    RunnableLambda(lambda x: f"其他类型: {type(x).__name__}")
)

# 测试
print(branch.invoke(42))       # 数字: 42
print(branch.invoke("hello"))  # 字符串: hello
print(branch.invoke([1,2,3]))  # 其他类型: list
```

相当于switch语句,满足其中一个条件执行

```python
print(branch.invoke("abcd"))  # 其他类型: list
```
